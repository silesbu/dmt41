## Task 1

We first load and preprocess the data before we do any analysis or training

```python
df = pd.read_csv("data/ODI-2018.csv")

df.head()
```

```python
df.columns = ["timestamp","programme","ml","ir","stats",
              "databases","gender","chocolate","birthday",
              "neighbors","stand up","earnings","random number",
              "bed time","good day 1","good day 2"]

# Cast nominal values to numeric representation
df["gender"].replace({"male":0,"female":1,"unknown":2},inplace=True)
df["ml"].replace({"yes":1,"no":0,"unknown":2},inplace=True)
df["ir"].replace({"0":0,"1":1,"unknown":2},inplace=True)
df["stats"].replace({"sigma":0,"mu":1,"unknown":2},inplace=True)
df["databases"].replace({"nee":0,"ja":1,"unknown":2},inplace=True)
df["chocolate"].replace({"slim":0,"fat":1,"neither":2,"I have no idea what you are talking about":3},inplace=True)
df["stand up"].replace({"no":0,"yes":1,"unknown":2},inplace=True)
```

### Distribution of random numbers

```python
randomNumbers = df["random number"][1:]

nums = []
fake = []
for number in randomNumbers:
    if not number.isdigit() or int(number) < 0 or int(number) > 10:
        fake.append(number)
    else:
        nums.append(int(number))

print("Number of poisoned instances: %i" % len(fake))
        
plt.figure(figsize=(10,5))
plt.hist(nums)
plt.show()
```
